```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io

img_orig = io.imread('https://yoyoyo-yo.github.io/Gasyori100knock/dataset/images/imori_256x256.png')
img_noise = io.imread('https://yoyoyo-yo.github.io/Gasyori100knock/dataset/images/imori_256x256_noise.png')
img_gray = cv2.cvtColor(img_orig, cv2.COLOR_RGB2GRAY)

plt.subplot(1, 3, 1); plt.imshow(img_orig); plt.title('orig')
plt.subplot(1, 3, 2); plt.imshow(img_noise); plt.title('noise')
plt.subplot(1, 3, 3); plt.imshow(img_gray, cmap='gray'); plt.title('gray')
plt.show()
```

```python
# Q.11 平滑化フィルタ, Smoothing filter

# opencv
img_smooth = cv2.blur(img_noise, (5, 5)) # smoothing filtering

plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.title('input')
plt.imshow(img_noise)
plt.subplot(1, 2, 2)
plt.title('smoothing fileter')
plt.imshow(img_smooth)
plt.show();
```

```python
def filter_smooth(img, ksize=(3,3)):
    _img = img.copy().astype(np.float32)
    ksize_h, ksize_w = ksize
    
    # padding
    h, w = img.shape[:2]
    pad_top, pad_bottom = ksize_h, ksize_h
    pad_left, pad_right = ksize_w, ksize_w
    
    _img = np.pad(_img, [(pad_top, pad_bottom), (pad_left, pad_right), (0, 0)], 'edge')
    out = np.zeros_like(_img)
    
    new_h, new_w = out.shape[:2]
    c = 1 if len(out.shape) == 2 else out.shape[2]
    
    # filtering
    for iy in range(new_h - ksize_h):
        for ix in range(new_w - ksize_w):
            for ic in range(c):
                out[iy, ix, ic] = np.mean(_img[iy : iy + ksize_h, ix : ix + ksize_w, ic])
    out = out[pad_top:pad_top + h, pad_left:pad_left + w]
    return np.clip(out, 0, 255).astype(np.uint8)
        
```

```python
img_smooth = filter_smooth(img_smooth, (5, 5))
plt.figure(figsize = (12, 3))
plt.subplot(1, 2, 1)
plt.title('input')
plt.imshow(img_noise)
plt.subplot(1, 2, 2)
plt.title('smooth filter')
plt.imshow(img_smooth)
plt.show();
```

```python
# カーネルサイズを大きくするとノイズが消えると同時に画像がぼやける
img_smooth = filter_smooth(img_noise, (9, 9))
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.title('input')
plt.imshow(img_noise)
plt.subplot(1, 2, 2)
plt.title('kernel size = 9')
plt.imshow(img_smooth)
plt.show()
```

```python
# Q.12 モーションフィルタ, Motion filter
# モーションフィルタとは対角方向の平均値を取るフィルタ

def filter_motion_cv2(img, ksize = (3, 3)):
    kernel = np.zeros(ksize)
    kernel[range(ksize[0]), range(ksize[0])] = 1/ksize[0]
    return cv2.filter2D(img, -1, kernel)

img_motion = filter_motion_cv2(img_orig, (5, 5))
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.title('input')
plt.imshow(img_orig)
plt.subplot(1, 2, 2)
plt.title('motion filter')
plt.imshow(img_motion)
plt.show();
```

```python
def filter_motion(img, ksize=(3, 3)):
    _img = img.copy().astype(np.float32)
    ksize_h, ksize_w = ksize
    
    # padding
    h, w = img.shape[:2]
    pad_top, pad_bottom = ksize_h, ksize_h
    pad_left, pad_right = ksize_w, ksize_w
    
    _img = np.pad(_img, [(pad_top, pad_bottom), (pad_left, pad_right), (0, 0)], 'edge')
    out = np.zeros_like(_img)
    
    new_h, new_w = out.shape[:2]
    c = 1 if len(out.shape) == 2 else out.shape[2]
    
    # define kernel
    k = np.zeros(ksize)
    k[range(ksize[0]), range(ksize[0])] = 1/ksize[0]
    
    # filtering
    for iy in range(new_h - ksize_h):
        for ix in range(new_w - ksize_w):
            for ic in range(c):
                out[iy, ix, ic] = np.sum(_img[iy : iy + ksize_h, ix:ix + ksize_w, ic] * k)
        
    out = out[pad_top: pad_top + h, pad_left: pad_left + w]
    return np.clip(out, 0, 255).astype(np.uint8)
```

```python
img_motion = filter_motion(img_orig, (5, 5))
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.title('input')
plt.imshow(img_orig)
plt.subplot(1, 2, 2)
plt.title('filter motion')
plt.imshow(img_motion)
plt.show();
```

```python
# Q.13 Max-Minフィルタ, Max-Min filter
def filter_max_min(img,  ksize = (3, 3)):
    _img = img.copy().astype(np.float32)
    ksize_h, ksize_w = ksize
    
    # padding
    h, w = img.shape[:2]
    pad_top, pad_bottom = ksize_h, ksize_h
    pad_left, pad_right = ksize_w, ksize_w
    
    if len(_img.shape) == 2:
        # -1が最後の次元に対応し、後ろからの位置を指定できる。
        _img = np.expand_dims(_img, axis = -1)
    
    _img = np.pad(_img, [(pad_top, pad_bottom), (pad_left, pad_right), (0, 0)], 'edge')
    out = np.zeros_like(_img)
    
    new_h, new_w = out.shape[:2]
    c = 1 if len(out.shape) == 2 else out.shape[2]
    
    # filtering
    for iy in range(new_h - ksize_h):
        for ix in range(new_w - ksize_w):
            for ic in range(c):
                out[iy, ix, ic] = np.max(_img[iy:iy + ksize_h, ix:ix + ksize_w, ic]) - np.min(_img[iy:iy + ksize_h, ix:ix + ksize_w, ic])
    
    out = out[pad_top: pad_top + h, pad_left: pad_left + w]
    return np.clip(out, 0, 255).astype(np.uint8)
```

```python
img_mm = filter_max_min(img_gray, (3, 3))
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.title('input gray')
plt.imshow(img_gray, cmap='gray')
plt.subplot(1,2 ,2)
plt.title('max_min')
plt.imshow(img_mm, cmap='gray')
plt.show();
```

```python
# Q.14. 微分フィルタ, Differential filte
# 微分フィルタは輝度の急激な変化が起こっている部分のエッジを取り出すフィルタ
_w = img_orig.shape[1]
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_orig)
plt.plot(np.arange(_w), [150] * _w, 'red')
plt.subplot(1, 3, 2)
plt.plot(np.arange(_w), img_orig[150, :, 0])
plt.show();
```

```python
def filter_diff(img, mode='y'):
    kernel= np.array([[0, 0, 0], [-1, 1, 0], [0, 0, 0]])
    if mode == 'y':
        kernel = np.array([[0, -1, 0], [0, 1, 0], [0, 0, 0]])
    return cv2.filter2D(img, -1, kernel)
```

```python
img_diff_x = filter_diff(img_gray)
img_diff_y = filter_diff(img_gray, mode = 'y')

plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)
plt.title('input')
plt.imshow(img_gray, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('x')
plt.imshow(img_diff_x, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('y')
plt.imshow(img_diff_y, cmap='gray')
plt.show();
```

```python
# Q.15. プリューウィットフィルタ, Prewitt filter
def filter_prewitt(img, ksize=(3, 3), mode = 'x'):
    k = np.zeros(ksize)
    if mode == 'x':
        k[:, 0] = 1
        k[:, -1] = -1
    else:
        k[0, :] = 1
        k[-1, :] = -1
    
    return cv2.filter2D(img, -1, k)
```

```python
img_pre_x = filter_prewitt(img_gray)
img_pre_y = filter_prewitt(img_gray, mode='y')

plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)
plt.title('input')
plt.imshow(img_gray, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('x')
plt.imshow(img_pre_x, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('y')
plt.imshow(img_pre_y,cmap='gray')
plt.show();
```
